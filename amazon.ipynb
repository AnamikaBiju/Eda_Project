#### About Dataset
The Amazon Prime Users Dataset contains information about 2500 fictional users of the Amazon Prime subscription service. Each entry in the dataset includes details such as the user's name, email address, location, subscription plan, payment information, and engagement metrics. Additionally, demographic data such as gender and date of birth are provided, along with user preferences such as favorite genres and devices used to access the platform.

The dataset aims to represent a diverse range of Prime users, including different demographics, subscription plans, and usage patterns. It is designed to facilitate analysis and insights into user behavior, preferences, and interactions with the Amazon Prime platform.

## Analysing favorite Genres on gender basis

```python
data = df
df = pd.DataFrame(data)
genre_counts = df.groupby('Gender')['Favorite_genres'].value_counts()
plt.figure(figsize=(10,25))
plt.subplot(1, 2, 1)
genre_counts['Male'].plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette("rainbow", len(genre_counts['Male'])))
plt.title('Male Favorite Genres')
plt.subplot(1, 2, 2)
genre_counts['Female'].plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette("rainbow", len(genre_counts['Female'])))
plt.title('Female Favorite Genres')
plt.tight_layout()
plt.show()
```

##### By analysing the favorite type of movie on the gender basis , we find that the largest slice for favorite male genres is Drama (14.7%) followed by Action (14.4%), Comedy (13.9%), and Romance (13.3%).Whereas , The largest slice for favorite female genres is also Comedy (14.1%) followed by Drama (14.5%), Romance (13.9%), and Action (12.4%).

##### Overall, Comedy and Drama are the most popular genres for both males and females.
